```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

diamondsSeaborn = sns.load_dataset('diamonds')

# Create a FacetGrid
seaborngraph = sns.FacetGrid(diamondsSeaborn, col="cut", col_wrap=3, height=5, hue="color")

# Function to plot scatter plot with jitter
def scatter_with_jitter(x, y, **kwargs):
    # Add small random noise to the data
    x_jittered = x + np.random.normal(0, 0.1, len(x))
    y_jittered = y + np.random.normal(0, 0.1, len(y))
    
    sns.scatterplot(x_jittered, y_jittered, alpha=0.7, marker="D", edgecolor="w", s=30, **kwargs)

# Map the scatter_with_jitter function to each subplot
seaborngraph.map(scatter_with_jitter, "price", "carat")

# Add legend
seaborngraph.add_legend()

# Show the plot
plt.show()
```
