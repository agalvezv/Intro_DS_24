1. [10%] FacetGrid in Seaborn is a class that helps in creating a grid of subplots based on the
values of one or more categorical variables. Using the appropriate seaborn or plotly
functions, generate a jittered scatter plot of the price of a diamond as a function of diamond
weight broken into five sub-panels across two rows, where each sub-panel represents a
different cut quality. Depict the points' color as the cut. (Extra 5 points, change the default
background color to white).

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamondsSeaborn = sns.load_dataset('diamonds')
seaborngraph = sns.FacetGrid(diamondsSeaborn, col="cut", col_wrap=3, height=4, hue="color",aspect=2.5)
seaborngraph.map(sns.scatterplot, "carat", "price",alpha=0.7, marker="D", edgecolor="w",s=30, x_jitter=.3, y_jitter=.3)
seaborngraph.add_legend()
plt.show()
```

4. [20%] Provide the code to generate the following plot, where the bands around the fitted
lines are 95% confidence intervals.
Provide code below:


```python
import seaborn as sns
import matplotlib.pyplot as plt

diamondsSeaborn = sns.load_dataset('diamonds')

seaborngraph = sns.FacetGrid(diamondsSeaborn, col="cut", col_wrap=3, height=4, aspect=2.5)
seaborngraph.map(sns.regplot, "carat", "price", scatter_kws={'alpha': 0.7, 'marker': "D", 'edgecolor': "w", 's': 30}, ci=95)
seaborngraph.add_legend()
plt.suptitle('Relationship between Carat and Price by Diamond Cut', y=1.1)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamondsSeaborn = sns.load_dataset('diamonds')
sns.regplot(x='carat',y='price',data=diamondsSeaborn, x_ci='ci', scatter=True, fit_reg=True, ci=95, x_jitter=.5, y_jitter=.5, marker='D')
plt.show()
```

7. [5%] What is a violin plot, and what function in seaborn can be used to generate one?

A violin plot is a graphical representation in data science and statistics that is a combination of a box plot and a kernel density plot.It can help tell a person about the shape of the data, the probability density at certain points due to its curves and also shows us similar details to a box plot. The median, quartiles and other aspects of the central tendency can be see using this specific graph format. The function seaborn.violinplot() can be used to generate a violin plot using seaborn.

8. [5%] How are violin plots different from box plots?

The advantages of using a violin plot is that is can provide a few features a box plot cannot. It is able to show distribution visualization as stated earlier, proving the probability distribution of the data. It provides users with information about the shape of the data and its modes.Another benefit is that is can display an estimate of how common certain values are within the information because of the density estimation factor. It can show multiple peaks or clusters when multimodal data is involved. The nature of the graph also provides information about differences between groups because of its shape. 

9. [10%] In the context of visualizing data distributions with seaborn boxplots, outliers are typically represented as individual points. However, when multiple outliers exist within the same category, they can overlap, potentially obscuring important details. How might we address this issue of overlapping outliers in seaborn boxplots to enhance the clarity of our visualizations? Consider the role of an additional seaborn plot that adds a jitter layer overlaid on a boxplot to improve the representation of individual data points. What specific function and parameters would you utilize to achieve this, and how would you incorporate it into the existing boxplot code provided below?

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamondsSeaborn = sns.load_dataset('diamonds')


sns.boxplot(x="cut", y="price", data=diamondsSeaborn)
sns.stripplot(x="cut", y="price", data=diamondsSeaborn, color="black", size=.4, jitter=True, alpha=0.5)

plt.show()
```

11.	[15%] Provide the code to generate the following plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamondsSeaborn = sns.load_dataset('diamonds')

sns.histplot(data=diamondsSeaborn, x='depth', hue='cut', multiple='stack', bins=30)

plt.xlabel('Depth(%)')
plt.ylabel('Count')
plt.title('Distribution of Diamond Depth by Cut')

cut_labels = diamondsSeaborn['cut'].unique()

plt.legend(title='Cut', labels=cut_labels, loc='upper right', edgecolor='black')

plt.show()
```
